PROPOSITIONAL LOGIC

A **proposition** is a declarative sentence that is either "True" or "False", but not both.


Example:


"Paris is the capital of France." is a *true proposition*.


"New York is the capital of the United States of America" is a *false proposition*.

"What time is it?" *is not a proposition,* as it is not a declarative sentence (it is a interogative one).

"x + 1 = 2" *is not a proposition* as it is not True or False *depending* on the *value of x*. It can be turned into a proposition if a value is *assigned* to x. This study area of the logic is called **"Propositional Calculus"**.

```python
"""
In Python, the truth value of a variable can be represented by the
Boolean Variable #True or #False.
"""
p = True
q = False
```

CONJUNCTION

The **conjunction** of two propositions p,q noted **"p and q"** or **"(p)(q)"** is *True when both p and q are True and False otherwise*.

```python
#conjunction
def conjuction(p,q):
  if p == True and q == True:
    r = True
  else:
    r = False
  return r
```

```python
# examples
conjuction(p,p)
```

```python
conjuction(p,q)
```

DISJUNCTION

The **disjunction** of two propositions p,q noted **"p or q"** or **"p + q"** *is False when both p and q are False  and  True otherwise*.

```python
# disjuction
def disjuction(p,q):
  if p == False and q == False:
    r = False
  else:
    r =  True
  return r
```

```python
#examples
disjuction(p,p)
```

```python
disjuction(q,q)
```

```python
#Inverter Gate
def inverter(x=off): #default value is off
  if x:
    y = off
  else:
    y = on
  return y


inverter()
```

# APPLICATION: TWO-SWITCHES CIRCUIT

Sometimes light fixtures are controlled by more than one switch. Circuits need to be designed so that flipping any of the switches for the fixture turns the light on when it is off and turns the light off when it is on. Design circuits that accomplish this when here are two switches.

  XNOR-equivalent-multi-768x320.webp

  Figure 6: Two-switches circuit: X-NOR Circuit

```python
# Two switches circuit
def two_switches_circuit(x,y):
  # xy
  x1 = and_gate(x,y)
  # (-x)
  x2 = inverter(x)
  # (-y)
  y2 = inverter(y)
  # (-x)(-y)
  y1 = and_gate(x2,y2)
  # xy + [(-x)(-y)]
  return or_gate(x1,y1)
```

```python
#examples
two_switches_circuit(x=on,y=on)
```

```python
two_switches_circuit(x=on,y=off)
```

```python
two_switches_circuit(x=off,y=on)
```

```python
two_switches_circuit(x=off,y=off)
```
